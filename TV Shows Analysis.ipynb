## The watch counts according to years. 

```python
watch_year_count = data['Year'].value_counts()
print("Most Watched Years:\n", watch_year_count)

# Visualization
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
watch_year_count.plot(kind='bar', color='skyblue')
plt.title("Most Watched Years")
plt.xlabel("Year")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

### 2020 was the year I watched the most number of TV Show episodes, probably because of the pandemic restrictions. There is not a specific pattern for the other years. 

## The watch counts according to months. 

```python
watch_month_count = data['Month'].value_counts()
print("Most Watched Months:\n", watch_month_count)

# Visualization
plt.figure(figsize=(10, 6))
watch_month_count.plot(kind='bar', color='orange')
plt.title("Most Watched Months")
plt.xlabel("Month")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

## The most number of episodes watched by the TV Show title. 

```python
# Count occurrences of each TV show
most_watched_shows = data['Title'].value_counts()

# Display top 10 most-watched shows
print("Top 10 Most Watched TV Shows:")
print(most_watched_shows.head(10))

# Visualization
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 8))
most_watched_shows.head(10).plot(kind='bar', color='skyblue')
plt.title("Top 10 Most Watched TV Shows")
plt.xlabel("TV Show Title")
plt.ylabel("Watch Count")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

### Friends, Modern Family, How I Met Your Mother, Full House, Brooklyn Nine Nine and That 70s Show are all popular sitcoms that have lasted for around 10 seasons each. I religiously watched them regardless of how long they are and clearly enjoy sitcoms. 

## Are there rewatches of the Same Episode? 

```python
# Create a unique identifier for Title, Season, and Episode
data['UniqueID'] = data['Title'] + " | Season: " + data['Season'].astype(str) + " | Episode: " + data['Episode'].astype(str)

# Count the number of times each unique identifier has been watched
rewatched_counts = data['UniqueID'].value_counts()

# Filter for rewatches (watched more than once)
rewatched = rewatched_counts[rewatched_counts > 1]

# Display rewatched items
print("Rewatched TV Shows (Title | Season | Episode):")
print(rewatched)

# Visualization
if not rewatched.empty:
    plt.figure(figsize=(10, 6))
    rewatched.sort_values(ascending=False).plot(kind='bar', color='cyan')
    plt.title("Rewatched TV Shows")
    plt.xlabel("Show | Season | Episode")
    plt.ylabel("Rewatch Count")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.show()
else:
    print("No TV shows have been rewatched.")
```

## I have watched some of the Full House episodes twice and some other episodes randomly. 

## Content According to the Seasons

```python
# Define seasons based on months
season_mapping = {
    "December": "Winter", "January": "Winter", "February": "Winter",
    "March": "Spring", "April": "Spring", "May": "Spring",
    "June": "Summer", "July": "Summer", "August": "Summer",
    "September": "Fall", "October": "Fall", "November": "Fall"
}
data['Season'] = data['Month'].map(season_mapping)

# Count occurrences of each season
most_watched_season = data['Season'].value_counts()

# Display most-watched season
print("Most Watched Seasons:")
print(most_watched_season)

# Visualization
plt.figure(figsize=(8, 6))
most_watched_season.plot(kind='bar', color='teal')
plt.title("Content Watched by Season")
plt.xlabel("Season")
plt.ylabel("Watch Count")
plt.tight_layout()
plt.show()
```

## Fall being the most popular isn't surprising, however the season after Fall, Winter being the least popular is surprising. Why the sudden drop?

## Content According to Episode Names

```python
# Filter out rows where the 'Episode' column contains the word 'Episode'
filtered_data = data[~data['Episode'].str.contains("Episode", na=False)]

# Count occurrences of each remaining episode
most_watched_episodes = filtered_data['Episode'].value_counts()

# Display most-watched episodes
print("Most Watched Episodes (Filtered):")
print(most_watched_episodes.head(10))

# Visualization
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
most_watched_episodes.head(10).plot(kind='bar', color='pink')
plt.title("Top 10 Most Watched Episodes (Filtered)")
plt.xlabel("Episode")
plt.ylabel("Watch Count")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

### Pilot must be filtered as well since it's a general term. 

```python
# Filter out rows where the 'Episode' column contains the word 'Episode' or 'Pilot'
filtered_data = data[
    ~data['Episode'].str.contains("Episode", na=False) &
    ~data['Episode'].str.contains("Pilot", na=False)
]

# Count occurrences of each remaining episode
most_watched_episodes = filtered_data['Episode'].value_counts()

# Display most-watched episodes
print("Most Watched Episodes (Filtered, excluding 'Episode' and 'Pilot'):")
print(most_watched_episodes.head(10))

# Visualization
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
most_watched_episodes.head(20).plot(kind='bar', color='pink')
plt.title("Top 10 Most Watched Episodes (Filtered, Excluding 'Episode' and 'Pilot')")
plt.xlabel("Episode")
plt.ylabel("Watch Count")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
